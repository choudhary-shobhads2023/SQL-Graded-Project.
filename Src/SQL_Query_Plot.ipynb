**[Q1] What is the distribution of customers across states?**

```python
sql_Res.head()
```

**[Q8] What is the trend of revenue and orders by quarters?**

```python
sql_Res8 = pd.read_csv("Result_set8.csv")
```

```python
sql_Res8.head()
```

```python
fig, ax1 = plt.subplots(figsize=(8,5))
# using the twinx() for creating another axes object for secondary y-Axis
# Instantiate a second axes that shares the same x-axis
ax2 = ax1.twinx()
# ax2.set_ylim(4, 20)
ax1.set_title("Trend of Revenue and orders by quarters",color='orange')
ax1.plot(sql_Res8.quarter_number,sql_Res8.Revenue,color='green',marker='o',label='Revenue')
ax2.plot(sql_Res8.quarter_number,sql_Res8.Number_of_orders,color='blue',marker='*',label='Count of Orders')
ax1.set_xlabel("Quarter Number",color='orange')
ax1.set_ylabel("Revenue",color='orange')
ax2.set_ylabel("Number Of Orders",color='orange')
# plt.xticks(color='blue')
# plt.yticks(color='blue')
# plt.yticks(color='blue')
ax1.legend(["Revenue"])
ax2.legend(["Count of Orders"],loc='upper center')
plt.tight_layout()
plt.show()
```

**Insights:**

* We observed a decreasing Trend in both Revenue and Number_of_orders quarter by quarter.
* Orders in quarter 4 is not generating much revenue.

**[Q9] What is the average discount offered for different types of credit cards?**

```python
sql_Res9 = pd.read_csv("Result_set9.csv")
```

```python
sql_Res9.head()
```

```python
# plt.bar(sql_Res9.credit_card_type,sql_Res9.Average_discount)
plt.figure(figsize=(10,5))
ax = sns.barplot(data=sql_Res9,x='credit_card_type',y='Average_discount',palette='crest')
# Iterating over the bars one-by-one
for bar in ax.patches:
    # Using Matplotlib's annotate function and
  # passing the coordinates where the annotation shall be done
  # x-coordinate: bar.get_x() + bar.get_width() / 2
  # y-coordinate: bar.get_height()
  # free space to be left to make graph pleasing: (0, 8)
  # ha and va stand for the horizontal and vertical alignment
    ax.annotate(format(bar.get_height(), '.2f'), 
                   (bar.get_x() + bar.get_width() / 2, 
                    bar.get_height()), ha='center', va='center',
                   size=8, xytext=(0, 8),
                   textcoords='offset points',color='blue')
    
plt.title("Average discount for different types of credit cards",color='green')
plt.xlabel("Credit Card Types",color='green')
plt.ylabel("Average Discount offered",color='blue')
plt.xticks(rotation=90,color='blue')
plt.yticks(color='blue')
plt.show()
```

**Insights:**

* Average discount for different Credit cards Ranges from 0.58 to 0.64
* Laser credit card offer higher discount of 0.64
